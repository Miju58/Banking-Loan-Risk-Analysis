```python
df[['credit_score','credit_risk']].groupby('credit_risk').mean()
```

```python
df['credit_risk'].value_counts().plot(kind='bar')

plt.title('Credit Risk Distribution')
plt.xlabel('Risk (0=Low Risk, 1=High Risk)')
plt.ylabel('Customer Count')

plt.show()
```

```python
df.groupby('credit_risk')['credit_score'].mean().plot(kind='bar')

plt.title('Average Credit Score by Risk')
plt.ylabel('Average Credit Score')

plt.show()
```

```python
pd.crosstab(
    df['loan_default_history'],
    df['credit_risk']
)
```

```python
pd.crosstab(
    df['fraud_flag'],
    df['credit_risk']
)
```

```python
df.groupby('credit_risk')['total_outstanding_debt'].mean()
```

```python
df.groupby('credit_risk')['total_outstanding_debt'].mean().plot(kind='bar')

plt.title('Average Outstanding Debt by Risk')
plt.ylabel('Debt Amount')

plt.show()
```

```python
# Employment Status vs Risk
pd.crosstab(
    df['employment_status'],
    df['credit_risk']
)
```

```python
# Average Income by Risk
df.groupby('credit_risk')['annual_income'].mean()
```

```python
# Average Loan Amount by Risk
df.groupby('credit_risk')['loan_application_amount'].mean()
```

```python
df['risk_category'] = df['credit_risk'].map({
    0: 'Low Risk',
    1: 'High Risk'
})

df.to_csv('C:/Users/mijan/OneDrive/Desktop/Banking-Loan-Risk-Analysis/Dataset/credit_risk_dataset.csv', index=False)
```
